 > El algoritmo de Bernstein funciona de manera similar al de Deutsch-Jozsa, pero en vez de tratar de distinguir entre clases de funciones, busca el string de una cadena de caracteres compuesta de ceros y unos que caracteriza a la funci√≥n dada.

```python
from qiskit import QuantumCircuit, transpile
from qiskit import Aer, execute
from qiskit.visualization import plot_histogram 
```

<p style="background-color:salmon; color:white; padding: 33px"> Empezaremos preguntandole a nuestra cadena que valor toma en cualquier cadena que quedaramos.

```python
Circuit = QuantumCircuit(4)
Circuit.x(2)
Circuit.ccx(1, 2, 3)
Circuit.x(2)
Circuit.ccx(1, 2, 3)
Circuit.draw(output='mpl')
```

```python
from qiskit import QuantumCircuit,transpile
```

```python
result = transpile(Circuit, basis_gates=['u3', 'cx'])
result.draw(output='mpl')
```

```python
costo = result.count_ops()['u3'] + 10*result.count_ops()['cx']
print(costo)
```

```python
orac1 = Circuit.to_gate()
orac1.name = "BERSTEIN VAZIRANI"
```

```python
a = QuantumCircuit(4)
a.append(orac1, [0,1,2,3])
a.draw(output = "mpl")
```

```python
Circuit = QuantumCircuit(4,3)
Circuit.x(3)
Circuit.h(3)
Circuit.h(0)
Circuit.h(1)
Circuit.h(2)
Circuit.append(orac1, range(4))
Circuit.h(0)
Circuit.h(1)
Circuit.h(2)
Circuit.measure(0,0)
Circuit.measure(1,1)
Circuit.measure(2,2)

Circuit.draw(output='mpl')
```

```python
simulador = Aer.get_backend('qasm_simulator')
```

```python
ejecucion = execute(Circuit, backend=simulador, shots=1) 
resultado = ejecucion.result()
conteos = resultado.get_counts()
plot_histogram(conteos)
```
